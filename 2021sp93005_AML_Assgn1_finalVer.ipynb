Let us write a small function to check for null values in the dataset for both train and test.

```python
def missingvalues(data):
    tot = data.isnull().sum().sort_values(ascending = False)
    per = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    ms=pd.concat([tot, per], axis=1, keys=['Total', 'Percentage'])
    ms= ms[ms["Percentage"] > 0]
    f,ax =plt.subplots(figsize=(8,6))
    plt.xticks(rotation='90')
    fig=sns.barplot(ms.index, ms["Percentage"],color="purple",alpha=0.8)
    plt.xlabel('Features', fontsize=15)
    plt.ylabel('Percent of missing values', fontsize=15)
    plt.title('Percent missing data by feature', fontsize=15)
    return ms
```

```python
#For making the data ready let us first combine both the data (train and test) to 
#one single entity so that we can apply certain functions on them together
whole_data=[train_df, test_df]

```

```python
whole_data
```

```python
#as sibsp is siblings / spouses aboard the Titanic and parch is parents / children aboard the Titanic.
# we combine them into one feature called family
for data in whole_data:
    data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
```

```python
import re
#Let us now extract titles from passenger names
def title_cal(name):
    t_search = re.search(' ([A-Za-z]+)\.', name)
    if t_search:
        return t_search.group(1)
    return ""
for data in whole_data:
    data['Title'] = data['Name'].apply(title_cal)
for data in whole_data:
    data['Title'] = data['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    data['Title'] = data['Title'].replace('Mlle', 'Miss')
    data['Title'] = data['Title'].replace('Ms', 'Miss')
    data['Title'] = data['Title'].replace('Mme', 'Mrs')
```

```python
## creating bins for age features
for data in whole_data:
    data['Age_bin'] = pd.cut(data['Age'], bins=[0,12,20,40,120], labels=['Children','Teenager','Adult','Senior'])
```

```python
## creating bins for fare features
for data in whole_data:
    data['Fare_bin'] = pd.cut(data['Fare'], bins=[0,7.91,14.45,31,120], labels=['Low_fare','median_fare',
                                                                                      'Average_fare','high_fare'])
```

```python
traindf=train_df
testdf=test_df
all_d=[traindf,testdf]
```

```python
for data in all_d:
    drop_col = ['Age','Fare','Name','Ticket']
    data.drop(drop_col, axis=1, inplace = True)
```

```python
drop_col = ['PassengerId']
traindf.drop(drop_col, axis=1, inplace = True)
```

```python
testdf.head()
```

```python
traindf.head()
```

```python
testdf = pd.get_dummies(testdf, columns = ["Sex","Title","Age_bin","Embarked","Fare_bin"],
                             prefix=["Sex","Title","Age_type","Em_type","Fare_type"])
```

```python
traindf = pd.get_dummies(traindf, columns = ["Sex","Title","Age_bin","Embarked","Fare_bin"],
                             prefix=["Sex","Title","Age_type","Em_type","Fare_type"])
```

**Splitting the data**

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score  
from sklearn.model_selection import KFold 
from sklearn.model_selection import cross_val_score 
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
features = traindf.drop("Survived",axis=1)
Target_feature = traindf["Survived"]
X_train,X_test,y_train,y_test = train_test_split(features,Target_feature,test_size=0.3,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# Implement the Machine learning Techniques:

**Model 1: Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
prediction=model.predict(X_test)
print('Accuracy for LR is ',round(accuracy_score(prediction,y_test)*100,2))
kfold = KFold(n_splits=10, random_state=22 ,shuffle=True)     #Splitting the data into 10 equal parts
result_LR=cross_val_score(model,features,Target_feature,cv=10,scoring='accuracy')
print('The cross validated score for LR is:',round(result_LR.mean()*100,2))
y_pred = cross_val_predict(model,features,Target_feature,cv=10)
sns.heatmap(confusion_matrix(Target_feature,y_pred),annot=True,fmt='3.0f',cmap="OrRd")
plt.title('Confusion Matrix', y=2, size=20)
```

**Model 2: Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(criterion='gini', n_estimators=700,
                             min_samples_split=10,min_samples_leaf=1,
                             max_features='auto',oob_score=True,
                             random_state=1,n_jobs=-1)
model.fit(X_train,y_train)
prediction=model.predict(X_test)
print('Accuracy for RF is',round(accuracy_score(prediction,y_test)*100,2))
kfold = KFold(n_splits=10, random_state=22, shuffle= True) # k=10, split the data into 10 equal parts
result_RFC=cross_val_score(model,features,Target_feature,cv=10,scoring='accuracy')
print('The cross validated score for Random Forest Classifier is:',round(result_RFC.mean()*100,2))
y_pred = cross_val_predict(model,features,Target_feature,cv=10)
sns.heatmap(confusion_matrix(Target_feature,y_pred),annot=True,fmt='3.0f',cmap="PuRd")
plt.title('Confusion Matrix', y=2, size=20)
```

```python
from sklearn.svm import SVC, LinearSVC
model = SVC()
model.fit(X_train,y_train)
prediction=model.predict(X_test)
print('The accuracy of the Support Vector Machines Classifier is',round(accuracy_score(prediction,y_test)*100,2))
kfold = KFold(n_splits=10, random_state=22, shuffle= True) # k=10, split the data into 10 equal parts
result_SVM=cross_val_score(model,features,Target_feature,cv=10,scoring='accuracy')
print('The cross validated score for Support Vector Machines Classifier is:',round(result_SVM.mean()*100,2))
y_pred = cross_val_predict(model,features,Target_feature,cv=10)
sns.heatmap(confusion_matrix(Target_feature,y_pred),annot=True,fmt='3.0f',cmap="YlGn")
plt.title('Confusion_matrix', y=2, size=20)
```

**HYPERPARAMTER TUNING**

Tuning is done only for Random Forest Classifier

```python
train_X = traindf.drop("Survived", axis=1)
train_Y=traindf["Survived"]
test_X  = testdf.drop("PassengerId", axis=1).copy()
train_X.shape, train_Y.shape, test_X.shape
```

```python
from sklearn.model_selection import GridSearchCV
model = RandomForestClassifier()
n_estim=range(100,1000,100)
## Search grid for optimal parameters
param_grid = {'criterion':['gini','entropy'],'max_depth':np.arange(1, 50), 'min_samples_leaf': range(1, 10)}
model_rf = GridSearchCV(model,param_grid = param_grid, cv=5,scoring="accuracy", n_jobs= 4, verbose = 1)
model_rf.fit(train_X,train_Y)
# Best score
print("Best Score: ", model_rf.best_score_)
print("Best parameters: ", model_rf.best_params_)
model_rf.best_params_
print("Best estimator: ", model_rf.best_estimator_)
```

```python

random_forest = RandomForestClassifier(criterion='entropy', max_depth=20, min_samples_leaf=2)
random_forest.fit(train_X, train_Y)
Y_pred_rf = random_forest.predict(test_X)
result_RF_HYP=random_forest.score(train_X,train_Y)
print('Accuracy for Random Forest after hyper parameter tuning is',random_forest.score(train_X,train_Y)*100)



```
